Perfect  lets pin down the **syntax of `.loc` and `.iloc`** clearly.

---

#  `.loc`  **Label-based selection**

```python
df.loc[row_labels, column_labels]
```

* `row_labels`  a single label, list of labels, slice, or condition
* `column_labels`  a single label, list, or slice

 Examples:

```python
df.loc["row1", "col1"]          # single value
df.loc["row1"]                  # single row
df.loc[:, "col1"]               # single column
df.loc["row1":"row3", "col1":"col2"]  # label-based slice (inclusive)
df.loc[df["Age"] > 30, "Name"]  # condition-based filter
```

---

#  `.iloc`  **Integer position-based selection**

```python
df.iloc[row_positions, column_positions]
```

* `row_positions`  integer index (0, 1, 2), list, or slice
* `column_positions`  integer index (0, 1, 2), list, or slice

 Examples:

```python
df.iloc[0, 1]        # single value (row 0, col 1)
df.iloc[0]           # first row
df.iloc[:, 1]        # second column
df.iloc[0:2, 0:2]    # rows 01, cols 01 (end exclusive)
df.iloc[[0, 2], [1]] # specific rows and columns
```

---

#  Key Rule of Thumb

* `.loc`  **think names/labels**
* `.iloc`  **think numbers/positions**

Perfect   lets dive into **broadcasting in Pandas DataFrames**.

Because Pandas is built on NumPy, it inherits **broadcasting** but adds **label alignment** on top.

---

#  1. Scalar Broadcasting

A scalar value is **broadcast** to every element in the DataFrame:

```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

print(df + 10)
```

Output:

```
    A   B
0  11  14
1  12  15
2  13  16
```

---

#  2. Series Broadcasting (by Column Index)

If you add a **Series** with matching column labels, values broadcast **down each column**:

```python
s = pd.Series([100, 200], index=["A", "B"])
print(df + s)
```

Output:

```
     A    B
0  101  204
1  102  205
2  103  206
```

---

#  3. Series Broadcasting (by Row Index)

If you want to broadcast by **row**, set `axis=0`:

```python
s_row = pd.Series([10, 20, 30], index=[0, 1, 2])
print(df.add(s_row, axis=0))
```

Output:

```
    A   B
0  11  14
1  22  25
2  33  36
```

---

#  4. DataFrame with Different Shapes

Pandas tries to **align on both index and columns**.
If labels dont match  result is `NaN`.

```python
df2 = pd.DataFrame({"A": [1, 2, 3]}, index=[0, 1, 2])
print(df + df2)
```

Output:

```
     A   B
0   2 NaN
1   4 NaN
2   6 NaN
```

---

#  5. NumPy-Style Broadcasting

Since under the hood its NumPy, you can also do shape-based broadcasting:

```python
import numpy as np

print(df.values + np.array([10, 20]))   # row-wise broadcast
```

---

 **Summary**:

* Scalars  apply to all elements.
* Series  broadcast across rows or columns (depending on index alignment + `axis`).
* DataFrames  align on both **rows and columns**.
* If labels dont match  you get `NaN`.

Got it   `.dropna()` in **Pandas** is used to **remove missing values (NaN / None)** from a **Series** or **DataFrame**.

---

#  1. On a Series

```python
import pandas as pd
import numpy as np

s = pd.Series([1, np.nan, 2, None, 3])

print(s.dropna())
```

Output:

```
0    1.0
2    2.0
4    3.0
dtype: float64
```

 All NaN/None values are dropped.

---

#  2. On a DataFrame

```python
df = pd.DataFrame({
    "A": [1, np.nan, 3],
    "B": [4, 5, np.nan]
})

print(df.dropna())
```

Output:

```
     A    B
0  1.0  4.0
```

 Only the row with **no NaN values** is kept.

---

#  3. Important Parameters

* **`axis`**  where to drop NaNs

  * `axis=0` (default)  drop rows
  * `axis=1`  drop columns

```python
df.dropna(axis=1)   # drops columns with any NaN
```

* **`how`**  when to drop

  * `"any"` (default)  drop if *any* NaN present
  * `"all"`  drop if *all* values are NaN

```python
df.dropna(how="all")
```

* **`thresh`**  minimum number of non-NaN values required

```python
df.dropna(thresh=1)   # keep rows with at least 1 non-NaN
```

* **`subset`**  only check specific columns

```python
df.dropna(subset=["A"])   # drop rows where A is NaN
```

* **`inplace`**  modify DataFrame directly

```python
df.dropna(inplace=True)
```

---

 **Summary**:

* `.dropna()` removes rows/columns with missing values.
* Flexible with `axis`, `how`, `thresh`, and `subset`.
* Use `inplace=True` to modify without creating a copy.

## Exploratory Data Analysis with Pandas ##

Perfect   **Exploratory Data Analysis (EDA)** with **Pandas** is all about **understanding, cleaning, and summarizing** your dataset before applying machine learning or deeper analytics.

Heres a step-by-step **EDA workflow in Pandas** (with examples).

---

#  1. Import Libraries & Load Data

```python
import pandas as pd

df = pd.read_csv("data.csv")   # or create DataFrame
```

---

#  2. Get Basic Info

```python
print(df.head())       # first 5 rows
print(df.tail())       # last 5 rows
print(df.shape)        # (rows, columns)
print(df.info())       # data types + missing values
print(df.describe())   # summary stats (mean, std, etc.)
```

---

#  3. Columns & Data Types

```python
print(df.columns)      # list of column names
print(df.dtypes)       # data types
```

Convert if needed:

```python
df["Age"] = df["Age"].astype(int)
```

---

#  4. Missing Values

```python
print(df.isna().sum())         # count missing values
df = df.dropna()               # drop missing rows
# OR
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())  # fill with mean
```

---

#  5. Duplicates

```python
df.duplicated().sum()     # check
df = df.drop_duplicates() # remove
```

---

#  6. Univariate Analysis

```python
print(df["Age"].value_counts())   # frequency of ages
print(df["Salary"].mean())        # average salary
print(df["Salary"].median())      # median salary
print(df["Salary"].std())         # standard deviation
```

---

#  7. Filtering & Conditions

```python
high_salary = df[df["Salary"] > 70000]
print(high_salary)
```

---

#  8. Grouping & Aggregation

```python
print(df.groupby("Age")["Salary"].mean())   # avg salary by age
```

---

#  9. Correlation

```python
print(df.corr(numeric_only=True))   # correlation matrix
```

---

#  10. Visualization (using Pandas built-in)

```python
df["Age"].hist()                 # histogram
df.boxplot(column="Salary")      # boxplot
df.plot.scatter(x="Age", y="Salary")  # scatter plot
```

---

 **Summary of EDA with Pandas**:

1. Inspect dataset (`head`, `info`, `describe`).
2. Handle missing & duplicate values.
3. Explore distributions (`value_counts`, `mean`, `median`).
4. Filter & group for insights.
5. Check correlations.
6. Visualize trends.

## SQL in pandas ##

Great one   you can **use SQL-like operations directly in Pandas**, because Pandas was designed with SQL-style data manipulation in mind.

---

#  1. SELECT  Column Selection

```python
import pandas as pd

df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35]
})

# SQL: SELECT name, age FROM df;
print(df[["name", "age"]])
```

---

#  2. WHERE  Filtering Rows

```python
# SQL: SELECT * FROM df WHERE age > 30;
print(df[df["age"] > 30])
```

---

#  3. ORDER BY  Sorting

```python
# SQL: SELECT * FROM df ORDER BY age DESC;
print(df.sort_values("age", ascending=False))
```

---

#  4. GROUP BY + Aggregation

```python
# SQL: SELECT age, COUNT(*) FROM df GROUP BY age;
print(df.groupby("age").size())

# SQL: SELECT age, AVG(id) FROM df GROUP BY age;
print(df.groupby("age")["id"].mean())
```

---

#  5. JOINs

```python
df2 = pd.DataFrame({
    "id": [1, 2, 4],
    "salary": [50000, 60000, 70000]
})

# SQL: SELECT * FROM df INNER JOIN df2 ON df.id = df2.id;
print(pd.merge(df, df2, on="id", how="inner"))
```

Types of joins in Pandas:

* `"inner"`  INNER JOIN
* `"left"`  LEFT JOIN
* `"right"`  RIGHT JOIN
* `"outer"`  FULL OUTER JOIN

---

#  6. LIKE  String Filtering

```python
# SQL: SELECT * FROM df WHERE name LIKE 'A%';
print(df[df["name"].str.startswith("A")])
```

---

#  7. IN Operator

```python
# SQL: SELECT * FROM df WHERE age IN (25, 35);
print(df[df["age"].isin([25, 35])])
```

---

#  8. Using Actual SQL Queries on DataFrames

You can also install **pandasql** to run raw SQL:

```python
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())

q = "SELECT name, age FROM df WHERE age > 28"
print(pysqldf(q))
```

---

 **Summary**:

* Pandas already supports SQL-like operations (`select`, `where`, `groupby`, `join`, etc.).
* For real SQL queries, use **pandasql** or **duckdb**.

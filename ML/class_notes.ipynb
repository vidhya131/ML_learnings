## Perceptron?

A **Perceptron model** is one of the simplest types of artificial neural networks, introduced by **Frank Rosenblatt in 1958**. It’s a **binary linear classifier**, meaning it tries to separate data into two classes (e.g., yes/no, 0/1, positive/negative) using a linear decision boundary.

---

### Structure of a Perceptron

A perceptron has:

1. **Inputs**: $x_1, x_2, ..., x_n$ (features of the data).

2. **Weights**: $w_1, w_2, ..., w_n$ (importance of each feature).

3. **Bias**: $b$ (offset that helps shift the decision boundary).

4. **Summation function**: Computes the weighted sum:

   $$
   z = \sum_{i=1}^{n} w_i x_i + b
   $$

5. **Activation function**: Applies a step function to decide the output:

   $$
   y =
   \begin{cases}
   1 & \text{if } z \geq 0 \\
   0 & \text{if } z < 0
   \end{cases}
   $$

---

### Working

* The perceptron takes inputs, multiplies them by weights, adds a bias, and passes the result through an activation function (step function).
* During training, the weights and bias are adjusted to minimize misclassification using the **perceptron learning rule**.

---

### Limitations

* A single-layer perceptron can only solve **linearly separable problems** (like AND, OR).
* It **cannot solve non-linear problems** (like XOR).
* This limitation led to the development of **multilayer perceptrons (MLPs)** and deeper neural networks.

---

✅ In short:
The perceptron is the **foundation of neural networks**, acting as a simple linear classifier that inspired modern deep learning.

Great question! Let’s connect the **Perceptron**, **Machine Learning (ML)**, and **Linear Regression** step by step:

---

## 🔹 Perceptron and Machine Learning

* The perceptron is an **early supervised learning algorithm** for binary classification.
* It’s a simple **linear classifier**: it learns a weight vector $w$ and bias $b$ to separate data into two classes with a linear decision boundary (a line in 2D, a plane in 3D, a hyperplane in higher dimensions).
* This idea—**learning weights from data**—is the foundation of modern ML, especially neural networks.

---

## 🔹 Perceptron vs. Linear Regression

Both use a **linear function** of inputs:

* **Linear Regression model**:

  $$
  \hat{y} = w \cdot x + b
  $$

  Here, the goal is to fit a line that minimizes the squared error between predictions and true outputs (continuous values).

* **Perceptron model**:

  $$
  \hat{y} = \text{step}(w \cdot x + b)
  $$

  Instead of predicting continuous values, it applies a **step function** to classify inputs into two categories (0 or 1).

---

## 🔹 Key Differences

| Aspect        | Linear Regression                  | Perceptron                                       |
| ------------- | ---------------------------------- | ------------------------------------------------ |
| Output        | Continuous value                   | Discrete class (0/1)                             |
| Loss function | Mean Squared Error (MSE)           | Misclassification error (implicit)               |
| Learning      | Gradient descent on squared error  | Perceptron update rule                           |
| Use case      | Predict numbers (regression tasks) | Classify into two classes (classification tasks) |

---

## 🔹 Relationship

* You can think of the perceptron as **linear regression followed by a threshold (step function)**.
* In fact, logistic regression (a core ML model) came later as a **smoother version** of perceptron: instead of a hard step, it uses the sigmoid function for probabilities.

---

👉 So:

* **Linear regression** → fits a line to predict numbers.
* **Perceptron** → fits a line to separate classes.
* Both are foundational ML models based on linear functions.

## 🔹 What is Linear Regression?

Linear regression is one of the **simplest and most widely used statistical models**.
It tries to explain the relationship between a **dependent variable (Y)** and one or more **independent variables (X’s)** using a straight line (or hyperplane in higher dimensions).

---

## 🔹 Equation

For **simple linear regression** (one predictor):

$$
Y = \beta_0 + \beta_1X + \epsilon
$$

* $Y$: dependent variable (target)
* $X$: independent variable (feature)
* $\beta_0$: intercept (value of Y when X=0)
* $\beta_1$: slope (change in Y for a unit change in X)
* $\epsilon$: error term (captures what the line can’t explain)

For **multiple linear regression** (many predictors):

$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_nX_n + \epsilon
$$

---

## 🔹 How it works

* It estimates coefficients ($\beta$’s) using **Ordinary Least Squares (OLS)**, i.e. minimizing the sum of squared residuals:

  $$
  \text{SSE} = \sum (Y_i - \hat{Y}_i)^2
  $$
* Produces a best-fit line that predicts Y from X’s.

---

## 🔹 Assumptions

For linear regression to be reliable:

1. **Linearity** → relationship between X and Y is linear.
2. **Independence** → errors are independent.
3. **Homoscedasticity** → errors have constant variance.
4. **Normality** → errors are normally distributed.
5. **No multicollinearity** → independent variables aren’t highly correlated with each other.

---

## 🔹 Types

1. **Simple Linear Regression** → one X, one Y.
2. **Multiple Linear Regression** → multiple X’s, one Y.
3. **Polynomial Regression** → X raised to higher powers (captures curves, still linear in coefficients).
4. **Regularized Regression** → adds penalty terms:

   * Ridge (L2 penalty)
   * Lasso (L1 penalty)
   * ElasticNet (mix of both)

---

## 🔹 Example in Python

```python
import pandas as pd
import statsmodels.api as sm

# Example dataset
data = pd.DataFrame({
    "X": [1, 2, 3, 4, 5],
    "Y": [2, 4, 5, 4, 5]
})

# Prepare data
X = sm.add_constant(data["X"])  # adds intercept
y = data["Y"]

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Summary
print(model.summary())
```

This gives you coefficients, R², p-values, etc.

---

## 🔹 Strengths

* Easy to interpret (coefficients show effect size).
* Computationally efficient.
* Works well when assumptions hold.

---

## 🔹 Limitations

* Only models **linear relationships**.
* Sensitive to **outliers**.
* Can perform poorly if predictors are highly correlated (multicollinearity).
* Doesn’t handle complex, nonlinear relationships as well as ML models.

---

✅ In short: **Linear Regression is the go-to model for understanding relationships between variables and making predictions when the relationship is roughly linear.**

## 🔹 General Algorithm for Linear Regression (Gradient Descent)

1. Initialize parameters $\theta_0, \theta_1$.
2. Compute output $f(x)$ and the cost function $J$.
3. Update parameters using **gradient descent**:

   $$
   \theta_j = \theta_j - \alpha \frac{\partial J}{\partial \theta_j}, \quad j = 0,1
   $$

   where $\alpha$ = learning rate.
4. Repeat until $J$ is minimized (converges).

---

## 🔹 Example (Car Generation vs Price)

* Training data:

  ```
  Generation (X) | Price (Y in 1000s)
  1              | 1
  2              | 2
  3              | 4
  ```
* Start with $\theta_0 = 0.5, \theta_1 = 0.5, \alpha = 0.1$.
* First iteration: compute outputs, errors, gradients, then update parameters.
* After update, the model predicts better (errors are smaller).

👉 This shows how **gradient descent iteratively improves the fit**.

---

## 🔹 Regression Recap

* Model:

  $$
  f_\theta(x) = \theta_0 + \theta_1x
  $$
* Training data = pairs $(x_i, y_i)$.
* Applications: predicting **house prices, stock prices, exam scores, weather**.

---

## 🔹 Multi-variable Regression

* Instead of one input $x$, we can use many features:

  $$
  f_\theta(x) = w_0 + w_1 d_1 + w_2 d_2 + \dots + w_m d_m
  $$
* Gradient descent still applies, but now updates **all weights simultaneously**.

---

## 🔹 Algorithm Intuition

* If predictions are too **high**, reduce weights.
* If predictions are too **low**, increase weights.
* Updates depend on feature values and errors.

---

## 🔹 Working Example (House Rent Prediction)

* Features: **Size, Floor, Broadband**.
* Start with random weights.
* Compute prediction, error, squared error, gradient contributions.
* Update weights step by step.

---

## 🔹 Handling Categorical Features

* Convert categorical data (like “Yes/No”, “City A/B/C”) into numerical form (e.g., one-hot encoding).

---

## 🔹 Feature Scaling

* Normalize features to have mean = 0 and standard deviation = 1 (or scaled by range).
* Ensures all features contribute equally and gradient descent converges faster.

---

## 🔹 Learning Rate

* Too small → slow convergence.
* Too large → may not converge (oscillates).
* Solution: dynamic learning rate (decrease with iterations).

---

## 🔹 Bias-Variance Trade-off

* **High bias** → model is too simple (underfits).
* **High variance** → model too complex (overfits).
* Need balance.

---

## 🔹 Preventing Overfitting

* Reduce features.
* Control model complexity.
* Use **regularization** (L1 = Lasso, L2 = Ridge).
* Dropout (in neural networks).
* Increase batch size.

---

## 🔹 Regularization

* **Lasso (L1 penalty)** → encourages sparsity (feature selection).
* **Ridge (L2 penalty)** → discourages large weights (weight decay).

---

**R² (R-squared)** is a **statistical measure** used mainly in **regression analysis**.

### 🔹 Definition

* R² measures how well the independent variables (features) explain the variability of the dependent variable (target).
* It’s also called the **Coefficient of Determination**.

### 🔹 Formula

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

Where:

* $SS_{res}$ = Sum of squared residuals (errors)
* $SS_{tot}$ = Total sum of squares (variation in actual data)

### 🔹 Interpretation

* **R² = 1 (100%)** → Perfect fit (model explains all variability in data).
* **R² = 0** → Model explains nothing; just predicts the mean.
* **0 < R² < 1** → Model explains part of the variability (higher is better).
* **R² < 0** → Model is worse than predicting the mean (bad model).

### 🔹 Example

If R² = 0.85 → 85% of the variation in the target variable is explained by the predictors.

---

⚠️ **Note:**

* R² only works well for **linear regression** (with continuous targets).
* High R² doesn’t always mean a “good” model (it may be overfitting). That’s why **Adjusted R²** is also used when multiple predictors are present.

# 📘 Part 1: Lecture-Style Notes (Study Guide)

### 🔹 1. Linear Models and Motivation

* Suppose we want to **decide credit approval** based on features like age, salary, debt.
* Each feature (x_i) gets a weight (w_i).
* Compute a **linear score**:
  [
  s = w_0 + w_1x_1 + w_2x_2 + \dots + w_dx_d
  ]
* Apply a decision rule:

  * If (s > 0) → Approve (class = +1).
  * If (s < 0) → Deny (class = -1).

⚠️ Problem: This only gives **hard yes/no decisions**, not probabilities.

---

### 🔹 2. Logistic Regression Intuition

* Instead of a raw decision, we want **probabilities** (e.g., “70% chance of default”).
* Logistic regression transforms the linear signal (s = w^Tx) into a probability:
  [
  h(x) = \theta(s) = \frac{1}{1 + e^{-s}} \in [0,1]
  ]
* (h(x)) = probability that (y = 1) given features (x).
* Still uses **linear weights**, but outputs probabilities.

---

### 🔹 3. Training Logistic Regression

* Data: ((x_1, y_1), (x_2, y_2), \dots, (x_N, y_N)) with (y \in {0,1}).
* Model prediction:
  [
  \hat{y} = h(x) = \frac{1}{1 + e^{-w^Tx}}
  ]
* Loss function: **Cross-Entropy Loss**
  [
  L(w) = - \sum_{n=1}^N \Big( y_n \log(h(x_n)) + (1-y_n)\log(1-h(x_n)) \Big)
  ]
* Optimize using **Gradient Descent**:

  * Gradient:
    [
    \nabla L(w) = \sum_{n=1}^N (h(x_n) - y_n) x_n
    ]
  * Update rule:
    [
    w := w - \eta \cdot (h(x) - y) x
    ]
  * where (\eta) = learning rate.

---

### 🔹 4. Extensions

* **Non-linear features:** Transform inputs (e.g., (x^2), interactions) to capture curved relationships.
* **Multi-class classification:** Use **Softmax Regression**:
  [
  P(y=j|x) = \frac{e^{w_j^T x}}{\sum_k e^{w_k^T x}}
  ]
* Loss function: **Cross-Entropy for multi-class**.

---

### 🔹 5. Summary

* Logistic regression = linear model + sigmoid activation.
* Outputs probabilities in [0,1].
* Trained with cross-entropy loss + gradient descent.
* Flexible: can handle nonlinear features and multiple classes.

# 🧮 Part 2: Worked Example with Numbers

### Problem: Credit Approval (binary classification)

We have **two features**:

* (x_1 =) salary (in $1000s)
* (x_2 =) years employed

We want to predict:

* (y = 1) (approve credit)
* (y = 0) (deny credit)

---

### Step 1: Initialize Model

Let’s choose some weights:
[
w_0 = -4, \quad w_1 = 0.05, \quad w_2 = 1.0
]

So model is:
[
s = w_0 + w_1 x_1 + w_2 x_2
]
[
h(x) = \frac{1}{1+e^{-s}}
]

---

### Step 2: Sample Person

* Salary = 60,000 → (x_1 = 60)
* Years employed = 2 → (x_2 = 2)

Compute score:
[
s = -4 + (0.05)(60) + (1.0)(2)
= -4 + 3 + 2 = 1
]

---

### Step 3: Apply Logistic Function

[
h(x) = \frac{1}{1+e^{-1}} \approx 0.731
]

Interpretation:

* Model predicts **73% probability** that this person gets credit approved.
* If we use threshold = 0.5 → Approve.

---

### Step 4: Gradient Update (if training)

Suppose true label is (y=1).

* Prediction = 0.731
* Error = (0.731 - 1) = -0.269

Gradient update (learning rate (\eta = 0.1)):
[
w := w - 0.1 \cdot (h(x) - y) \cdot x
]
[
w_1 := 0.05 - 0.1(-0.269)(60) = 0.05 + 1.614 = 1.664
]
[
w_2 := 1.0 - 0.1(-0.269)(2) = 1.0 + 0.0538 = 1.054
]
[
w_0 := -4 - 0.1(-0.269)(1) = -4 + 0.0269 = -3.973
]

👉 After training, weights adjust to better fit the data.

---

✅ This shows how **logistic regression takes features → applies weights → predicts probability → updates weights** until convergence.

## Supervised learning ##

Supervised Learning Overview:

Supervised learning involves algorithms that learn from input-output pairs, where the correct output (label) is provided for each input.
The learning algorithm uses these examples to make predictions on new, unseen inputs.

Applications of Supervised Learning:

Examples include spam detection in emails, speech recognition, machine translation, online advertising, self-driving cars, and visual inspection in manufacturing.
Each application involves training the model with labeled data to predict outcomes for new inputs.

Types of Supervised Learning:

The content introduces two main types of supervised learning: 
                     regression and classification.
                     
Regression predicts continuous values (e.g., housing prices), while classification categorizes inputs into discrete classes (e.g., spam or not spam).




Supervised Learning Overview:

Supervised learning algorithms learn to predict the mapping from input (X) to output (Y).
Regression algorithms predict continuous numerical values, while classification algorithms predict discrete categories.

Classification Algorithms:

An example of classification is breast cancer detection, where the algorithm determines if a tumor is benign (0) or malignant (1).
Classification predicts a limited set of categories, unlike regression, which predicts any number.


Input Variables in Classification:

Classification can involve multiple input variables, such as tumor size and patient age.
The learning algorithm finds a boundary to separate different categories, aiding in diagnosis.
In summary, supervised learning encompasses both regression and classification, with classification focusing on predicting categories from a finite set of outputs.



## Unsupervised learning ##

Unsupervised learning is a key concept in machine learning that focuses on finding patterns in data without labeled outputs. 

Clustering Algorithms

Unsupervised learning involves grouping unlabeled data into clusters based on similarities.
An example is Google News, which groups related articles by identifying common words without human intervention.
Applications in Genetic Data

Clustering can also be applied to genetic data, where individuals are grouped based on gene activity without predefined categories.
This helps researchers identify different types of individuals based on genetic characteristics.
Market Segmentation

Companies can use clustering to segment customers into distinct groups for better service.
For instance, the DeepLearning.AI team identified different motivations among learners in their community, aiding in tailored support.

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
